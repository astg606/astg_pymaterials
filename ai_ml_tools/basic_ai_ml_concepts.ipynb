## <font color="red">Some definitions</font> <a class="anchor" id="sec_def"></a>

### <font color="blue">What is Artificial Intelligence?<font>

> Artificial intelligence (AI) involves machines and computer systems mimicking human intelligence processes like learning, reasoning, and self-correction.

- Using algorithms, data, and computational power, AI can reason, learn, and problem-solve by recognizing patterns based on experience and making informed data-based decisions.
- The ultimate goal of AI is to create machines that can simulate human intelligence, including reasoning, problem-solving, and creativity.


NASA provides a definition of Artifitial Intelligence in:

[What is Artificial Intelligence?](https://www.nasa.gov/what-is-artificial-intelligence/) 

### <font color="blue">What is Machine Learning?<font>

Let us start with a couple of notable quotes:

>**Machine learning enables a machine to automatically learn from data, improve performance from experiences, and predict things without being explicitly programmed.** - Arthur Samuel (1959)

>**Machine learning is the study of computer algorithms that allow computer programs to automatically improve through experience.** - Tom M. Mitchell (1997)


Machine Learning (ML) is a system that can learn from example through self-improvement and without being explicitly coded by programmer.
- It makes computers modify or adapt their actions so that these actions get more accurate, where accuracy is measured by how well the chosen actions reflect the correct ones. 
- It learns how the existing input and output data are correlated and it writes a rule to derive outputs from new inputs.
- It uses techniques from Statistics, Mathematics and Computer Science to make computer programs learn from data to predict an output. 


### <font color="blue">What is Deep Learning?<font>

> Deep Learning (DL) is a specialized subset of ML that uses neural networks (with multiple layers) to analyze complex patterns in large datasets.

- DL significantly advances AI by enabling systems to process and interpret unstructured data with unprecedented accuracy.
- DL models are typically used for complex tasks that involve large amounts of data, like image and speech recognition, natural language processing, robotics, and autonomous driving.


In summary:

- __AI__: _Any technique that enables computers to mimic human behavior._
- __ML__: _Ability to learn without explicitely being programmed._
- __DL__: _Extract paterns from data using artificial neural networks._

## <font color="red">How does Machine Learning work?</font> <a class="anchor" id="sec_ml_wk"></a>

- ML starts with data.
   - The data is gathered and prepared to be used as training data, or the information the ML model will be trained on.
- We supply the data to the ML algorithm and let it train itself to find patterns or make predictions. 
- To make an accurate prediction, ML algorithm sees an example, and when given a similar example, it can figure out the outcome.
- ML learns through the discovery of patterns that is made thanks to the provided <b>data</b>.

### <font color="blue">Supervised learning </font>

A method in which we provide sample labeled data to the machine learning system in order to train it, and on that basis, it predicts the output.

```
    f(X) = Y
```

- Here the known input data is tagged with the correct output.
   - The alogorithm then uses this relationship to learn a mapping from inputs to outputs, making it possible to predict the output for new data.
- This type of learning includes regression, classification
   - Decision trees, Support Vector Machines, Neural Networks, ...

### <font color="blue">Unsupervised learning</font>

- The training is provided to the machine with the set of data that has not been labeled, classified, or categorized, and the algorithm needs to act on that data without any supervision.

```
   f(X) = X'
```

- Correct responses are not provided, but instead the algorithm tries to identify similarities between the inputs so that inputs that have something in common are grouped together. 
- This type of learning includes clustering, dimensionality reduction
   - K-Means, Expectation Maximization, Principal component analysis, ...

### <font color="blue">Reinforcement learning</font>
- A feedback-based learning method, in which a learning agent gets a reward for each right action and gets a penalty for each wrong action. The agent learns automatically with these feedbacks and improves its performance.

```
   f(S, A, T, R) = π
```

- Reinforcement learning is a method that is concerned with how software agents ought to take actions in an environment in order to maximize the notion of cumulative reward. 
   - The __agent__ learns to achieve a __goal__ in an uncertain, potentially complex __environment__. 
   - An artificial intelligence faces a game-like situation. The computer employs trial and error to come up with a solution to the problem. 
   - To get the machine to do what the programmer wants, the artificial intelligence gets either __rewards__ or __penalties__ for the __actions__ it performs. 
   - Its goal is to __maximize__ the total reward.
- __The computer employs a “rewards/punishments” system, offering feedback to the algorithm to learn from its own experiences by trial and error.__
- Reinforcement learning differs from supervised learning in not needing labelled input/output pairs be presented, and in not needing sub-optimal actions to be explicitly corrected. Instead the focus is on finding a balance between exploration (of uncharted territory) and exploitation (of current knowledge).

## <font color="red"> Issue with Underfitting and Overfitting</font> <a class="anchor" id="sec_ml_fitt"></a>

There are two general problems in ML:

1. The algorithm does not perform well on the training dataset.
2. The algorithm performs well on the training dataset, but fails to perform on the evaluation data or the new data, and it’s unable to model the given distribution. 
    
    
We want to have an algorithm that performs well with training data as well as new data. This is known as **generalization**. Our main goal is to have  algorithms that are good at generalization.
       

### <font color="blue">Underfitting</font>

A __model is underfitting__ the training data when the it performs poorly on the training data (training error is too high). 
    
- The model is unable to capture the relationship between the input examples and the target values.
- There are high prediction errors for both training and validation data. 
- Some reasons causing the problem:
    - The model might be too simple: the input features are not explanatory enough to describe the target well. 
    - Data used for training is not cleaned and contains noise (garbage values) in it.
    - The model has a high bias.
    - The size of the training dataset used is not enough.

### <font color="blue"> Overfitting </font>

A __model is overfitting__ the training data when it performs well on the training data but does not do well on the validation data (the validation error is high compared to the training error).

- The model is memorizing the data it has seen and is unable to generalize to unseen examples.
- There is a very low prediction error on training data, but a very high prediction error on validation data.
- Some reasons causing the problem:
   - The model is highly complex: the input feature combinations are in a large number and affect the model’s flexibility.
   - Data used for training is not cleaned and contains noise (garbage values) in it.
   - The model has a high variance.
   - The size of the training dataset used is not enough.

## <font color="red">Important Terms</font> <a class="anchor" id="sec_ml_terms"></a>

We list and define a set of words that will be used throughout the presentations in this course series.

- **Machine Learning** (ML): Computer model that can learn from and make decisions and predictions based on data.
- **Feature space** (input): Input space of the model, where the variables live.
- **Feature vector**: Each entry in the feature space is referred to as an n-dimensional feature vector, where `n` is the number of features that define a particular data point.
- **Label space** (output): The set of labels or target variables associated with each of the feature vectors make up the label space.
- **Training**: The process of determining the ideal parameters comprising a model.
- **Training set**: The subset of the dataset used to train a model.
- __Hyperparameters__: Model settings that are predefined (by users) and govern the training process. They are chosen before the training procedure starts. In the case of neural networks, they include:
  - __Learning rate__: Controls how much the model’s weights are updated during each training step.
  - __Dropout rate__: Percentage of neurons to be randomly dropped from the network during training to prevent overfitting.
  - __Number of hidden layers__: Specifies the number of layers between the input and output layers in the neural network.
  - __Batch size__: Defines how many training examples are used in each iteration (Iterations are the number of batches required to complete one epoch).
  - __Number of epochs__: Determines how many times the entire training dataset is passed through the model. Since epochs can get quite large, it is often divided into several smaller batches.
- **Validation set**: A subset of the dataset—disjoint from the training set—used in validation.
- **Validation**: A process used, as part of training, to evaluate the quality of a machine learning model using the validation set. Because the validation set is disjoint from the training set, validation helps ensure that the model’s performance generalizes beyond the training set.
- __Overfitting__: Happens when a model learns certain patterns in the training so well that it performs badly on the validation set. This implies that the model is exclusively applicable to the training and does not adapt to fresh data.
  - If for instance the number of epochs is too high, there is a risk of overfitting.
- __Underfitting__: Happens when the model performs badly on both training and validation data. 
  - If for instance the number of epochs is too low the model might be underfitted.
- **Prediction**: A model output when provided with an input (from feature space).
