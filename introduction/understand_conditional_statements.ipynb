# <font color="red">What are Conditional Statements? </font>

- Conditional statements are part of our everyday lives. As people, __we can think about situations and make decisions based on what we observe or know to be true__. 
    - Is it raining?
    - Do I have milk in the house?
    - Do I have a enough money to pay for new shoes?
- In computer programming, __conditional statements are used to instruct the computer on the decision to make when given some conditions__.
  - They allow a program to evaluate a condition as either `True` or `False` (a Boolean expression) and alter the execution path based on the result
- These decisions are made if and only if the pre-stated conditions are either true or false , depending on the functions the programmer has in mind.
- Conditional statements work by using Boolean expressions in the form of logical `if/else` and `then` statements evaluating if something is true or false.

**We use conditional statements when we want to execute a particular block of code when a given condition is true or false.**

## <font color="blue">Core conditionational statement types<font>

There are three core types:

- __If Statements__: Execute a code only if the condition is true.
- __If-Else Statements__: Execute one block if the condition is true, another block if false.
- __Else-If Statements__: Check multiple conditions in sequence; run code for the first true condition.

# <font color="red">The `if` statement</font>

Runs a block of code only if the specified condition is true.

---

````python
mm(
    """
    graph TD
        Start([Start]) --> Condition{Is condition True?}
    
        Condition -->|True| IfBlock[Execute 'if' code block]
        Condition -->|False| End([Continue program execution])
    
        IfBlock --> End
    
        style Condition fill:#ff9900,stroke:#333,stroke-width:2px
        style IfBlock fill:#99ccff,stroke:#333,stroke-width:1px
    """
)
````

```python
if <expr>:
    <statement>
```

In the form shown above:

- `<expr>` is an expression evaluated in Boolean context: can be either `True` or `False`.
- `<statement>` is a valid Python statement, which must be indented.
   - If `<expr>` evaluates to `True`, `<statement>` is executed.
   - If `<expr>` evaluates to `False`, `<statement>` is skipped frtom execution.

# <font color="red">The `if`-`else` statement</font>

Provides an alternate path when the condition evaluates to false.

---

````python
mm(
    """
    graph TD
        Start([Start]) --> Condition{Is condition True?}
    
        Condition -->|True| IfBlock[Execute 'if' code block]
        Condition -->|False| ElseBlock[Execute 'else' code block]
    
        IfBlock --> End([Continue program execution])
        ElseBlock --> End([Continue program execution])
    
        style Condition fill:#ff9900,stroke:#333,stroke-width:2px
        style IfBlock fill:#99ccff,stroke:#333,stroke-width:1px
        style ElseBlock fill:#99ccff,stroke:#333,stroke-width:1px
    """
)
````

```python
   if <expr>:
       <statement1> when the expression is true
   else:
       <statement2> when the expression is false    
```

__Note:__ You cannot have an __`else`__ statement without an __`if`__ statement.

## <font color="purple">Going Deeper: Python Ternary Operator</font>

- A compact, one-line shortcut for a basic `if-else` assignment.
- If we have an `if...else` block, we can use the ternary operator and write the `if...else` block in one line:

```python
   <statement if true> if <expr> else <statement if false>
```

#### <font color="orange">Example 3</font>

`print a if a > b otherwise, print b`

````python
a = 100
b = 200
print(f'A = {a}') if a > b else print(f'B = {b}')
````

# <font color="red">More choices: the `elif` statement</font>

Evaluates multiple conditions sequentially until one matches.

---

````python
mm(
    """
    graph TD
        Start([Start]) --> Cond1{Is Condition 1 True?}
    
        Cond1 -->|True| Block1[Execute 'if' code block]
        Cond1 -->|False| Cond2{Is Condition 2 True?}
    
        Cond2 -->|True| Block2[Execute 'elif' code block]
        Cond2 -->|False| Block3[Execute 'else' code block]
    
        Block1 --> End([Continue program execution])
        Block2 --> End([Continue program execution])
        Block3 --> End([Continue program execution])
    
        style Cond1 fill:#ff9900,stroke:#333,stroke-width:2px
        style Cond2 fill:#ff9900,stroke:#333,stroke-width:2px
        style Block1 fill:#99ccff,stroke:#333,stroke-width:1px
        style Block2 fill:#99ccff,stroke:#333,stroke-width:1px
        style Block3 fill:#99ccff,stroke:#333,stroke-width:1px
    """
)
````

```python
   if <expr1>:
       <statement1>
   elif <expr2>:
       <statement2>
   elif <expr3>:
       <statement3>
   # and so on
```

__Note:__ Notice that there is no __`else`__ statement above. It is not required, but the __`else`__ statement is availble to be used if all the above conditionasl evaluate to be False.

#### <font color="orange">Example 4</font>: Body Temperature

````python
# provide the temperature of the patient.
temp = float(input('Enter the temperature of the patient: '))

if temp > 105.8:
    print(f'Patient {temp} degrees) has a very high fever.')
elif temp > 103.1:
    print(f'Patient {temp} degrees) has a high fever.')
elif temp > 100.4:
    print(f'Patient {temp} degrees) has fever.')
elif temp > 97.9:
    print(f'Patient {temp} degrees) is fine.')
else:
    print(f'Patient {temp} degrees) has low temperature.')
````

# <font color='red'> Combining expressions</font>

````python
a = 5
b = 5
c = 5

if a != b and b != c and a != c:
    print('This is a scalene triangle')
elif a == b and b == c:
    print('This is an equilateral triangle')
else:
    print('This is an isosceles triangle')
````

Individual values can evaluate to either `True` or `False`.

This can be useful as when used in an `if` condition or as operand of the Boolean operations (and, or, not)

- Some given names:
    - Values that evaluate to `False` are considered **Falsy**.
    - Values that evaluate to `True` are considered **Truthy**.

### Falsy Values

| Sequences and Collections | Numbers | Constants |
| :--- | :--- | :--- | 
| Empty lists `list()` | Zero of any numeric type | `None` |
| Empty tuples `tuple()` | Integer: `0` | `False` |
| Empty dictionaries `dict()` | Float: `0.0` |  |
| Empty sets `set()` | Complex: `0j` | |
| Empty strings `""` |  | |
| Empty ranges `range(0)` | | |

### Truthy Values

- Non-empty sequences or collections (lists, tuples, strings, dictionaries, sets).
- Numeric values that are not zero.
- `True`

Using the truthy and falsy concepts, the follwing code:

```python
    if len(my_list) != 0:
        print("Not empty list!")

    if my_string != "":
        print("Not empty string!")      
```

can be written as:

```python
    if my_list:
        print("Not empty list!")

    if my_string:
        print("Not empty string!") 
```
