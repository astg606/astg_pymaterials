# <font color="red"> Modular programming</font>

+ Python supports modular programming: process of breaking down a large and complex programming task into smaller and more manageable subtasks/modules.
+ Modularity has a lot of advantages:
  + Reusability
  + Maintainability
  + Simplicity
+ Functions, modules and packages are all constructs in Python that promote code modularization.

# <font color="red"> What is a module?</font>
+ A module is nothing but a file, with a `.py` extension, containing a Python code.
   - That code can include variables, functions, classes, and runnable statements.
   - The name of the file (without the `.py` extension) becomes the name of the module.
+ A module can be seen as a "toolbox" or a "library book."
   - Instead of writing every single piece of code in one massive file, you can organize related code into separate modules and then just "check out" or "borrow" the specific tools you need.
+ A module is accessed from other modules using the `import module_name` statement. 

## <font color="blue">Types of module</font>

In general, you will use three categories of modules:

- __Built-in modules__:
   - Python comes with a massive "Standard Library" of modules right out of the box. You don't need to install them; you just import them.
   - __Examples__: `math`, `os`, `sys`, `random`, `datetime`.
- __External / Third-Party modules__:
   - These are modules written by other people that you download and install.
   - __Examples__: `numpy` (for manipulating multidimensional numerical data), `pandas` (for data analysis), `requests` (for downloading web pages).
- __Custom modules__:
   - The ones you write yourself to organize your own project.

# <font color="red"> Writing your own Python module</font>

**Step 1:** Create a `.py` File

````python
%%writefile sample_module.py
"""
This is module that defines a variable and a function.
It can be imported or run as a script.
"""

print(f"__name__ is set to: {__name__}")

x = 2

def my_func():
    '''
    This is a sample function contains in a module.
    '''
    print('Hi from within a module!')



if __name__ == '__main__':
    print("<> Running the module as a script! <>")
    print(f"The value of x is: {x}")
    my_func()
````

````python
%cat sample_module.py
````

**Step 2:** Import the Module and Use Its Defined Variables and Functions

````python
import sample_module
````

````python
print(dir(sample_module))
````

````python
print(sample_module.__doc__)
````

````python
print(sample_module.x)
````

````python
sample_module.my_func()
````

````python
print(sample_module.my_func.__doc__)
````

### Running a module as a Script

We need to have `__name__` followed by code statements at the bottom of the file.

```python
if __name__ == '__main__':
    # Python code
```

- Python assigns the name of the Python module into a special variable called `__name__`.
- Depending on whether you are executing the program through the command line or importing the module into another module, the assignment for `__name__` will vary.
   - If you invoke your module as a script, the variable `__name__` has the value of `__main__` and the conditional evaluates to `True`.
   + If the module is imported into another module, `__name__` is the name of the imported module and the conditional evaluates to `False`.
- The statement `if __name__ == '__main__':` allows you to write code that only runs when you execute the script directly, but not when you import that script as a module into another file.
   - It allows a single Python file to serve a double purpose:
      - It can be run as a __standalone script__ (executing the code in the `if` block), or
      - It can be safely imported as a __reusable module__ (ignoring the code in the `if` block).

````python
print(sample_module.__name__)
````

````python
# Run the module as a script
%run sample_module.py
````

# <font color="red">`.pyc` Files</font>

---

- When you run or import a standard `.py` file, Python translates your human-readable source code into a low-level format called __bytecode__.
- Python saves this bytecode inside a `.pyc` file (__caching__).
- By caching the bytecode on your storage drive, Python skips the compilation step the next time you run the script.
   - This drastically speeds up application startup times and module loading.
   - Note that it does not make the code run faster once it is loaded; it only reduces the setup overhead.
- You can import a module without the `.py` file if you have a `.pyc` file.
- If you delete a `.pyc` file, Python will seamlessly regenerate it the next time the code runs.
-  Python checks the timestamp and file size of your `.py` source file. If you modify your code, Python automatically overwrites the outdated `.pyc` file.
-  Python can execute or import them even if the original `.py` source code file has been completely removed.
  - A `.pyc` file matches the specific Python version that generated it.
  - Do not use `.pyc` files to hide your code. They can easily be reversed back into fully readable Python files using available open-source tools.
- `.pyc` files are automatically created inside a hidden folder named `__pycache__`.


__Warning:__ Don't lose your `.py` source code. `.pyc` files aren't human readable.
